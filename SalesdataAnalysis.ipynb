## SALES DATA ANALYSIS

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = {
    "Month":["Jan","Feb","March","April","May","June","July"],
    "Product":["Laptop","Tablet","Tablet","Laptop","Mouse","Mobile","Mouse"],
    "Sales":[50000,40000,30000,55000,45000,25000,15000]
}
```

```python
df = pd.DataFrame(data)
```

```python
product_sales = df.groupby("Product")["Sales"].sum()
```

```python
product_sales.plot(kind="bar")
plt.title("Product Wise Sales")
plt.xlabel("Product")
plt.ylabel("Sales")
plt.grid(True)
plt.show()
```

```python
monthly_sales = df.groupby("Month")["Sales"].sum()
```

```python
monthly_sales.plot(kind = "line",marker = "o",color = "purple",linestyle = "-")
plt.title("Monthly Wise Sales")
plt.xlabel("Month")
plt.ylabel("Sales")
plt.grid(False)
plt.show()
```

```python
product_sales.plot(kind="pie",autopct = "%1.1f%%")
plt.title("Product Sales Distribution")
plt.ylabel("")
plt.show()
```
